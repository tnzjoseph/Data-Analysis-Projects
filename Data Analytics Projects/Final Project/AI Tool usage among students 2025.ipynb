<h2 style="color:#1A2CA3; text-align:left;">Project Objective </h2>

<p style="font-size:16px;text-align: justify;">
    
The objective of this project is to analyze AI tool usage among Indian students in the year 2025 and understand how it influences their academic performance, study habits, and learning behavior. By examining patterns and trends in the dataset, the project aims to identify how frequently students use AI tools, for what purposes they use them, and whether this usage has a positive, negative, or neutral impact on their grades. The insights gained will help in understanding the role of AI in modern education and support better decisions for effective and responsible use of AI in learning.</p>

<h2 style="color:#1A2CA3; text-align:left;">Problem Definition </h2>

<p style="font-size:16px; text-align: justify;">

With the rapid proliferation of AI tools like ChatGPT, Gemini, and Copilot in 2023-2025, Indian higher education institutions face critical questions about how these technologies are being adopted by students, their impact on learning outcomes, and the need for institutional policies. However, there is limited data-driven understanding of actual usage patterns, effectiveness, and potential risks associated with AI tool adoption in the Indian academic context.</p>

<h4 style="color:#4B4376; text-align:left;"> 2.3.1 AI Tool Usage Level</h4>

```python
df['Usage_Level'] = df['Daily_Usage_Hours'].apply( 
    lambda x: 'High' if x >= 4 else 'Medium' if x >= 2 else 'Low'
)
df
```

```python
df['Trust_Level']=df['Trust_in_AI_Tools'].apply (
    lambda x: 'High' if x>=4 else 'Medium' if x>=2 else 'Low'
)
df
```

<h3 style="color:#0F828C; text-align:left;"> Overall Summary</h3>
<p style="font-size:14px;">
This study analyzes the adoption patterns, usage behaviour, and academic impact of AI tools among 3,614 Indian college students (16 variables) in 2025. The findings indicate that AI tools have become deeply integrated into students’ academic lives across multiple states, streams, and years of study.</p>
<h4 style="color:#0F828C; text-align:left;"> AI Tool Preference </h4>

- ChatGPT (859) is the most preferred tool, closely followed by Gemini (854) and Copilot (836).
- A significant number (757) use other AI platforms, showing diverse adoption beyond major brands.
- Bard and Claude show comparatively lower usage.</p>
<p style="color:#0F828C; text-align:left;"><b>Insight:</b> AI adoption is competitive and not dominated by a single platform.</p>

<h4 style="color:#0F828C; text-align:left;"> Year-wise Adoption </h4>

- Second-year students (966) use and trust AI tools the most.
- First-year students follow, while final-year students show comparatively lower usage.
<p style="color:#0F828C; text-align:left;"><b>Insight:</b> Mid-program students rely more on AI, possibly due to academic workload and project demands.</p>


<h4 style="color:#0F828C; text-align:left;"> Academic Impact </h4>

- Students reporting positive grade impact form the largest group and show higher usage levels.
- Negative impact is reported by only a small portion of students.
<p style="color:#0F828C; text-align:left;"><b>Insight:</b> AI tools are generally perceived as academically beneficial rather than harmful.</p>

<h4 style="color:#0F828C; text-align:left;">Professors’ Permission </h4>

- 1,886 students are not allowed to use AI tools.
- 1,728 students are allowed to use them.
<p style="color:#0F828C; text-align:left;"><b>Insight:</b> Institutional policies remain divided, despite high student adoption.

<h4 style="color:#0F828C; text-align:left;">Stream-wise Usage</h4>

- Science, Engineering, and Arts students show relatively higher average usage.
- Law show comparatively moderate usage.
<p style="color:#0F828C; text-align:left;"><b>Insight:</b> AI integration varies across academic disciplines.

<h4 style="color:#0F828C; text-align:left;">Geographic Spread</h4>

- Maharashtra, Uttar Pradesh, and Rajasthan show the highest participation.
- Representation is well distributed across multiple Indian states.

<p style="color:#0F828C; text-align:left;"><b>Insight:</b> AI adoption is geographically widespread across India.</p>

<h4 style="color:#0F828C; text-align:left;">Willingness to Pay</h4>

- Not willing: 1,811 students
<p style="color:#0F828C; text-align:left;"><b>Insight:</b> Students are almost equally divided on paid AI access, indicating price sensitivity.

<h4 style="color:#0F828C; text-align:left;">Conclusion</h4>

<p>AI tools have become mainstream academic support systems among Indian college students in 2025.

While students generally report positive academic impact, high trust, and widespread adoption, institutional permissions remain divided and willingness to pay is balanced.

This highlights the need for clear institutional policies, structured AI integration frameworks, and ethical usage guidelines in higher education.</p>
